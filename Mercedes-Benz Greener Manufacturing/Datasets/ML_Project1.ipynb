### 1. Importing required packages

```python
import pandas as pd
```

### 2. Getting and Exlporing the data

```python
training_data = pd.read_csv("train.csv")
```

```python
training_data.head()
```

```python
training_data.describe(include="all")
```

```python
training_data.info
```

### 3. Splitting the training data into features and labels as X and y respectively

```python
X = training_data.drop('y', axis=1)
y = training_data['y']
```

### Apply the label encoding on features

```python
from sklearn.preprocessing import LabelEncoder
X = X.apply(LabelEncoder().fit_transform)
X
```

### 4. Find the features which has zero variance and remove it

```python
print("shape of X before removing zero variance features ", X.shape)
X = X.loc[:, (X != X.iloc[0]).any()]
print("shape of X after removing zero variance features ", X.shape)
```

### 5. Check if there are any null values/unique values

```python
X.isna().sum()
```

```python
X.nunique
```

### 6. Splitting the data for training into train and test

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y , test_size =0.3)
```

## 7. Perform dimensionality reduction

### Standardizing data

```python
from sklearn.preprocessing import StandardScaler 

sc = StandardScaler()

X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)
```

```python
from sklearn.decomposition import PCA
pca = PCA()

X_train_pca = pca.fit_transform(X_train)
X_test_pca = pca.fit_transform(X_test)
```

```python
pca.explained_variance_ratio_
```

```python
pca = PCA(n_components=300)

X_train_pca = pca.fit_transform(X_train)
X_test_pca = pca.fit_transform(X_test)
```

### 8. Build and test using XGBoost 

```python
import xgboost as xgb
```

```python
#change the data to a Dmatrix 
D_train  = xgb.DMatrix(X_train_pca, label = y_train)
D_test = xgb.DMatrix(X_test_pca, label = y_test)
```

```python
param = {
       'eta' : 0.3,
    'max_depth' : 2
}
```

```python
model = xgb.train(param, D_train, 21)
```

```python
y_pred = model.predict(D_test)
```

### 9. Checking the accuracy of the model

```python
from sklearn.metrics import mean_squared_error
import numpy as np
print('RMSE value of testing dataset is:')
print(np.sqrt(mean_squared_error(y_test, y_pred)))
```

### 10. Preparing data to be tested

```python
# get the unseen test data
unseen_data = pd.read_csv("test.csv")
#Applying label encoding
unseen_data = unseen_data.apply(LabelEncoder().fit_transform)
# standardizing the data
unseen_data = sc.fit_transform(unseen_data)
# fit the test data to PCA model
unseen_data_pca = pca.fit_transform(unseen_data)
# change the data to a D-matrix
D_unseen_data = xgb. DMatrix(unseen_data_pca)
# predicting the unseen data
y_pred_unseen_data = model.predict(D_unseen_data)
y_pred_unseen_data
```
